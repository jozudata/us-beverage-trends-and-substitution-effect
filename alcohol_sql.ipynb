```python
import duckdb
import pandas as pd
print("🚀Libraries loaded successfully!")
print(f"-> duckdb versión: {duckdb.__version__}")
print(f"-> Pandas versión: {pd.__version__}")
```

```python
# LOAD DATA SET
df = pd.read_csv('alcohol.csv')
df= df.drop_duplicates()
print("File loaded successfully!")
df.columns.tolist()
```

```python
# Define the time frame for analysis
start_year = 2013
end_year = 2023

# --- TOP 10 BEER ETHANOL CONSUMING STATES ---
beer_ethanol_query = f"""
    SELECT 
        state_name, 
        ROUND(AVG(ethanol_beer_gallons_per_capita), 3) AS avg_beer_ethanol
    FROM df
    WHERE year BETWEEN {start_year} AND {end_year}
    GROUP BY state_name
    ORDER BY avg_beer_ethanol DESC
    LIMIT 10;
"""
top_10_beer_ethanol_states = duckdb.query(beer_ethanol_query).to_df()

# --- TOP 10 SPIRITS ETHANOL CONSUMING STATES ---
spirits_ethanol_query = f"""
    SELECT 
        state_name, 
        ROUND(AVG(ethanol_spirit_gallons_per_capita), 3) AS avg_spirit_ethanol
    FROM df
    WHERE year BETWEEN {start_year} AND {end_year}
    GROUP BY state_name
    ORDER BY avg_spirit_ethanol DESC
    LIMIT 10;
"""
top_10_spirits_ethanol_states = duckdb.query(spirits_ethanol_query).to_df()

# --- TOP 10 WINE ETHANOL CONSUMING STATES ---
wine_ethanol_query = f"""
    SELECT 
        state_name, 
        ROUND(AVG(ethanol_wine_gallons_per_capita), 3) AS avg_wine_ethanol
    FROM df
    WHERE year BETWEEN {start_year} AND {end_year}
    GROUP BY state_name
    ORDER BY avg_wine_ethanol DESC
    LIMIT 10;
"""
top_10_wine_ethanol_states = duckdb.query(wine_ethanol_query).to_df()
```

```python
#Beer reporting
print("🍺 TOP 10 STATES BY BEER ETHANOL CONSUMPTION PER CAPITA (2013-2023):")
print("-" * 65)
print(top_10_beer_ethanol_states.to_string(index=False))
print("\n" + "="*65 + "\n")
```

```python
#Spirits reporting
print("🍸 TOP 10 STATES BY SPIRIT ETHANOL CONSUMPTION PER CAPITA (2013-2023):")
print("-" * 65)
print(top_10_spirits_ethanol_states.to_string(index=False))
print("\n" + "="*65 + "\n")
```

```python
#Wine reporting
print("🍷 TOP 10 STATES BY WINE ETHANOL CONSUMPTION PER CAPITA (2013-2023):")
print("-" * 65)
print(top_10_wine_ethanol_states.to_string(index=False))
```
